## Basic Gradient Descent optimization

```python
x = torch.tensor([10, 20, 30], dtype=torch.float32)
y = torch.tensor([50, 100, 150], dtype=torch.float32)

z = w*x
loss = torch.nn.functional.mse_loss(z, y)
```

```python
w
```

```python
learning_rate = 0.01
num_iterations = 100

print(f"Learning Rate: {learning_rate}")
print(f"Number of Iterations: {num_iterations}")
```

```python
learning_rate = 0.0001 # Reduced learning rate
num_iterations = 200 # Increased iterations for better convergence

x1 = x.reshape(3,1)
w = torch.randn(1, requires_grad=True) # Re-initialize w

for i in range(num_iterations):
    # Recalculate z and loss
    z = x1@w
    loss = torch.nn.functional.mse_loss(z, y)

    # Perform backpropagation
    loss.backward()

    # Update w using gradient descent
    with torch.no_grad():
        w -= learning_rate * w.grad

    # Zero the gradients. Gradients by default add up; to prevent double-counting, we explicitly zero them at each iteration.
    w.grad.zero_()

    # Print iteration, loss, and w every 100 iterations
    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}: Loss = {loss.item():.4f}, w = {w.tolist()}")

print(f"\nFinal w: {w.tolist()}")
```

## nn.Model version

```python
import torch

device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
print(f"Using {device} device")
```

```python
from torch import nn

class BasicRegression(nn.Module):
  def __init__(self, *args, **kwargs) -> None:
      super().__init__(*args, **kwargs)
      self.linear_stack = nn.Sequential(
          nn.Linear(in_features=1, out_features=10), # Increased output features
          nn.ReLU(),
          nn.Linear(in_features=10, out_features=10), # Increased output features
          # nn.ReLU(),
          # nn.Linear(in_features=10, out_features=10), # Increased input features
          # nn.ReLU(),
          # nn.Linear(in_features=10, out_features=10), # Increased output features
          nn.ReLU(),
          nn.Linear(in_features=10, out_features=1) # Increased input features
      )

  def forward(self, x):
    return self.linear_stack(x)
```

```python
model = BasicRegression().to(device)
print(model)
```

```python
for param in model.parameters():
    print(type(param), param.size())
```

```python
# change datatype to float
x = torch.arange(-100,100).type(torch.float32)
x = x.reshape(-1, 1) # Reshape to (100, 1)
y = x**3

# Normalize x and y
x_max = x.max()
y_max = y.max()
x = x / x_max
y = y / y_max

x = x.to(device)
y = y.to(device)

learning_rate = 0.005 # Can try standard Adam LR again
num_iterations = 5000 # Keep high iterations
batch_size = 10

model = BasicRegression().to(device) # Re-initialize the model with new architecture
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate) # Changed optimizer to Adam

def train_loop(model, loss_fn, optimizer, x, y, num_iterations):
  model.train()

  for i in range(num_iterations):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)

    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if (i + 1) % 1000 == 0: # Print every 1000 iterations
        print(f"Iteration {i+1}: Loss = {loss.item():.4f}")


train_loop(model, loss_fn, optimizer, x, y, num_iterations)
```

```python
x_test = torch.tensor([0.5]).to(device)
o = model(x_test).to(device)
print(o, x_test**3)
```

```python
import matplotlib.pyplot as plt

# Set the model to evaluation mode
model.eval()

# Generate predictions for the entire range of x values
with torch.no_grad():
    y_pred_normalized = model(x)

# Denormalize predictions and actual y values
y_pred_denormalized = y_pred_normalized * y_max
y_actual_denormalized = y * y_max
x_denormalized = x * x_max

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(x_denormalized.cpu().numpy(), y_actual_denormalized.cpu().numpy(), label='Actual y = x*x*x', alpha=0.7, s=20)
plt.plot(x_denormalized.cpu().numpy(), y_pred_denormalized.cpu().numpy(), color='red', label='Model Predictions')

plt.title('Model Predictions vs. Actual Values (Denormalized)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
